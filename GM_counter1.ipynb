```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Dhirendra719/Dhirendra-/refs/heads/main/thickness_data_with_mean_minus_13.csv')
df.head(11)
```

```python
# prompt: Using dataframe df: mean_minus_13/30 make a new columan

# Calculate the new column
df['Count/sec'] = df['Mean_minus_13'] / 30

# Display the updated DataFrame
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(
    df["Thickness(mm)"], df["Count/sec"],
    marker='o', linestyle='-', color='#1f77b4', linewidth=2, markersize=8,
    markerfacecolor='white', markeredgewidth=2, label='Counts per second'
)

# Add labels and title
plt.title("Effect of Thickness on Radiation Count Rate", fontsize=16, fontweight='bold')
plt.xlabel("Thickness (mm)", fontsize=13)
plt.ylabel("Count/sec", fontsize=13)

# Add your name as credit (change 'Your Name' to your real name if you want)
plt.text(
    0.5, 0.02, '©Dhirendra Pd. Upadhyay', fontsize=10, color='gray',
    ha='right', va='bottom', transform=plt.gca().transAxes
)

# Grid and style
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
